```python
def Summation (Value1, Value2, Value3):
    """Addition of numbers"""
    result1 = Value1 + Value3
    print("The sum is:", result1)
    result2 = Value1 + Value2 + Value3
    print("The sum of all three values is : ", result2)
    


print(Summation(Value3 = 2, Value1 = 5, Value2 = 10))
```

```python
def Names (one ,two,*extra):
    print("The  first two names are :", one, "and ", two)
  
    print("The other names are : ", *extra)


print(Names("James", "John", "Phil", "Peter", "Jake"))
```

```python
def Names (**all):
    print("His name is :", all)
    print("His middle name is ", all["middle_name"])
    


print(Names(first_name = "James", middle_name = "John",last_name = "Phil", other_name = "Geoffrey"))
```

```python
summation = lambda arg1, arg2: arg1 + arg2;
print("First Lambda:",summation(20,70))
product = lambda arg1, arg2: arg1 * arg2
print("\nSecond Lambda:",product(20,70))
exponential = lambda arg1: arg1 * arg1
print("\nThird Lambda:",exponential(20))
```
