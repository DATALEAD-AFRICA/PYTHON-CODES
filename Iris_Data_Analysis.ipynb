```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X = iris_data.drop(['Species'], axis =1)
Y = iris_data['Species']

X_train, X_test, Y_train ,Y_test = train_test_split( X, Y ,test_size = 0.3,  random_state = 50)
```

RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier

RF_model = RandomForestClassifier(random_state=50)

RF_model.fit(X_train, Y_train)
        
pred = RF_model.predict(X_test)
        
        
print(f'Model: {RF_model}\n')
    
print("Classification Report")
print(classification_report(Y_test,pred), "\n\n")
        
   
RF_clf=confusion_matrix(Y_test, pred)
fig,ax=plt.subplots(figsize=(6,4))
print(f"Confusion Matrix, Model: {RF_model}\n\n")
sns.heatmap(RF_clf,annot=True, cmap="Blues")
```

DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier

DT_model = DecisionTreeClassifier(random_state=50)
        
DT_model.fit(X_train, Y_train)
        
pred = DT_model.predict(X_test)
        
        
print(f'Model: {DT_model}\n')
    
print("Classification Report")
print(classification_report(Y_test,pred), "\n\n")
        
   
DT_clf=confusion_matrix(Y_test, pred)
fig,ax=plt.subplots(figsize=(6,4))
print(f"Confusion Matrix for  Model: {DT_model}\n\n")
sns.heatmap(DT_clf,annot=True, cmap="Blues")
```
